```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Step 1: Define the date range and intervals
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 11, 30)
date_range = pd.date_range(start=start_date, end=end_date, freq='15T')  # 15-minute intervals

# Step 2: Define service groups, service names, and device types
service_groups = {
    "Gaming": ["Fortnite", "League of Legends", "Valorant"],
    "Social Media": ["Instagram", "Facebook", "LinkedIn", "Twitter", "Snapchat", "TikTok"],
    "Streaming": ["YouTube", "Netflix", "Hulu", "Disney+", "Prime Video"],
    "Shopping": ["Amazon", "eBay", "Etsy"],
    "Software": ["Gmail", "Outlook", "BlueJeans", "Slack", "WebEx", "GMeet", "Microsoft Teams"]
}

device_types = {
    "Gaming": ["TV", "Mobile"],
    "Social Media": ["Mobile", "Tablet"],
    "Streaming": ["TV", "Mobile"],
    "Shopping": ["Mobile", "Tablet", "PC"],
    "Software": ["PC"]
}

# Create a reverse mapping from service name to service group
service_name_to_group = {name: group for group, names in service_groups.items() for name in names}

# Step 3: Define users and multiple MAC addresses
users = ["user1"]
mac_addresses = [f"02:00:00:{random.randint(0x00, 0xFF):02x}:{random.randint(0x00, 0xFF):02x}:{random.randint(0x00, 0xFF):02x}" for _ in range(5)]  # 5 MAC addresses

# Step 4: Generate the dataset
data_adjusted = []

for timestamp in date_range:
    for user, mac_address in zip(users, random.choices(mac_addresses, k=1)):  # Randomly assign one MAC per timestamp
        hour = timestamp.hour
        minute = timestamp.minute
        day_of_week = timestamp.weekday()

        # Determine service usage
        if day_of_week < 5:  # Weekdays (Monday to Friday)
            if 7 <= hour < 7.5:  # Morning Social Media/Streaming (7:00 AM - 7:30 AM)
                service_name = random.choice(service_groups["Social Media"] + service_groups["Streaming"])
            elif 9 <= hour < 19:  # Daytime Software usage (9:00 AM - 7:00 PM)
                service_name = random.choice(service_groups["Software"])
            elif 20 <= hour < 21.5:  # Evening Streaming (8:00 PM - 9:30 PM)
                service_name = random.choice(service_groups["Streaming"])
            elif 19 <= hour < 22:  # Alternate Gaming/Streaming (7:00 PM - 10:00 PM)
                if day_of_week % 2 == 0:  # Even weekdays (Monday, Wednesday, Friday)
                    service_name = random.choice(service_groups["Gaming"])
                else:  # Odd weekdays (Tuesday, Thursday)
                    service_name = random.choice(service_groups["Streaming"])
            elif day_of_week == 4 and 23 <= hour < 23.5:  # Fridays (11:00 PM - 11:30 PM)
                service_name = random.choice(service_groups["Shopping"])  # Fallback to Shopping
            else:  # General fallback pattern
                service_name = random.choice(service_groups["Social Media"] + service_groups["Streaming"] + service_groups["Shopping"])
        else:  # Weekends (Saturday and Sunday)
            if 7 <= hour < 7.5:  # Morning Social Media/Streaming (7:00 AM - 7:30 AM)
                service_name = random.choice(service_groups["Social Media"] + service_groups["Streaming"])
            elif 9 <= hour < 19:  # Daytime Software usage (9:00 AM - 7:00 PM)
                service_name = random.choice(service_groups["Software"])
            elif 19 <= hour < 23:  # Exclusive Netflix usage (7:00 PM - 11:00 PM)
                service_name = "Netflix"
            else:  # General fallback for weekends
                service_name = random.choice(service_groups["Social Media"] + service_groups["Streaming"])

        # Determine service group and device type
        service_group = service_name_to_group[service_name]
        device_type = random.choice(device_types[service_group])

        # Generate random values for usage
        usage_minutes = random.randint(1, 60)  # Max 60 minutes per 15-minute interval
        usage_percentage = min(usage_minutes / 60 * 100, 100)  # Ensure logical calculation

        # Generate random values for network-related features
        signal_strength = round(random.uniform(-60, -40), 2)
        packet_loss_rate = round(random.uniform(0, 1), 4)
        latency = round(random.uniform(10, 50), 2)
        jitter_ms = round(random.uniform(0, 10), 2)
        traffic_spike = random.choice([0, 1])
        bandwidth_speed_per_sec_mbps = round(random.uniform(5, 15), 2)
        buffer_occupancy = round(random.uniform(0, 1), 4)

        # Append row to dataset
        data_adjusted.append([
            user, timestamp, mac_address, service_group, service_name, device_type, usage_minutes, usage_percentage,
            signal_strength, packet_loss_rate, latency, jitter_ms, traffic_spike,
            bandwidth_speed_per_sec_mbps, buffer_occupancy
        ])

# Step 5: Create DataFrame and save
df_balanced_adjusted = pd.DataFrame(data_adjusted, columns=[
    "user", "timestamp", "mac_address", "service_group", "service_name", "device_type", "usage_minutes", "usage_percentage",
    "signal_strength", "packet_loss_rate", "latency", "jitter_ms", "traffic_spike",
    "bandwidth_speed_per_sec_mbps", "buffer_occupancy"
])

# Save to CSV
df_balanced_adjusted.to_csv("netflix_single_user_balanced_pattern_v3.csv", index=False)
print("Dataset created and saved as 'netflix_single_user_balanced_pattern_v3.csv'")
```
