```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (assuming the dataframe is named df_mock)
# df_mock = pd.read_csv('your_dataset.csv')  # Uncomment if loading from a file

# Display an overview of the data
print("Dataset Overview:\n", df_mock.describe(include='all').T)

# Set plot style
plt.style.use('default')

# 1. Count of device types
plt.figure(figsize=(8, 5))
sns.countplot(data=df_mock, x='Device_Type', palette='viridis')
plt.title('Distribution of Device Types')
plt.xlabel('Device Type')
plt.ylabel('Count')
plt.show()

# 2. Count of application types
plt.figure(figsize=(8, 5))
sns.countplot(data=df_mock, x='Application_Type', palette='plasma')
plt.title('Distribution of Application Types')
plt.xlabel('Application Type')
plt.ylabel('Count')
plt.show()

# 3. Signal Strength Distribution per Application Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_mock, x='Application_Type', y='Signal_Strength_RSSI', palette='coolwarm')
plt.title('Signal Strength (RSSI) Distribution per Application Type')
plt.xlabel('Application Type')
plt.ylabel('Signal Strength (RSSI)')
plt.show()

# 4. Latency Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df_mock['Latency_ms'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Latency')
plt.xlabel('Latency (ms)')
plt.ylabel('Frequency')
plt.show()

# 5. Bandwidth Usage Distribution per Application Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_mock, x='Application_Type', y='Bandwidth_Usage_MBps', palette='muted')
plt.title('Bandwidth Usage per Application Type')
plt.xlabel('Application Type')
plt.ylabel('Bandwidth Usage (MBps)')
plt.show()

# 6. Packet Loss Rate vs Priority Score
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_mock, x='Packet_Loss_Rate', y='Priority_Score', hue='Application_Type', palette='cool')
plt.title('Packet Loss Rate vs Priority Score by Application Type')
plt.xlabel('Packet Loss Rate')
plt.ylabel('Priority Score')
plt.legend(title='Application Type')
plt.show()

# 7. Buffer Occupancy Distribution by Time of Day
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_mock, x='Time_of_Day', y='Buffer_Occupancy', palette='Set2')
plt.title('Buffer Occupancy by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Buffer Occupancy')
plt.show()

# 8. Correlation heatmap for numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(df_mock.corr(numeric_only=True), annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Heatmap of QoS Features')
plt.show()
```
